```python
import pandas as pd
import numpy as np


path = '/content/drive/MyDrive/Healthcare_Project/'

df_patients = pd.read_csv(path + 'patients.csv')
df_doctors = pd.read_csv(path + 'doctors.csv')
df_appointments = pd.read_csv(path + 'appointments.csv')
df_billing = pd.read_csv(path + 'billing.csv')
df_treatments = pd.read_csv(path + 'treatments.csv')

print("All 5 files loaded successfully!")
print("Billing Columns:", df_billing.columns.tolist())
print("Patients Columns:", df_patients.columns.tolist())
```

```python

df_billing['payment_method'] = df_billing['payment_method'].fillna('Unknown')


df_patients['date_of_birth'] = pd.to_datetime(df_patients['date_of_birth'])
df_billing['bill_date'] = pd.to_datetime(df_billing['bill_date'])


current_year = 2026
df_patients['age'] = current_year - df_patients['date_of_birth'].dt.year


df_billing['collection_risk'] = np.where(
    (df_billing['amount'] > 4000) & (df_billing['payment_status'].isin(['Pending', 'Failed'])),
    'High Risk',
    'Normal'
)

print("Data Cleaning aur Feature Engineering complete!")

print(df_billing['collection_risk'].value_counts())
```

```python

df_merged_billing = pd.merge(df_billing, df_patients, on='patient_id', how='inner')

insurance_analysis = df_merged_billing.groupby('insurance_provider').agg(
    total_patients=('patient_id', 'count'),
    total_revenue=('amount', 'sum'),
    average_bill=('amount', 'mean')
).reset_index()


insurance_analysis = insurance_analysis.sort_values(by='total_revenue', ascending=False)

print("--- FINANCIAL INSIGHTS BY INSURANCE  ---")
display(insurance_analysis)
```
